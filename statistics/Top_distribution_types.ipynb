```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import e
```

```python
# Симуляция равномерного распределения
uniform_dist = []

for _ in range(10):
    random_value = np.random.uniform(0.99, 1, 1)
    uniform_dist.append(random_value)

# Распределение результатов будет соответствовать равномерному распределению.
sns.displot(uniform_dist, bins=20)
plt.show()
```

```python
# Симуляция нормального распределения
normal_dist = []

for _ in range(10000):
    random_value = np.random.normal()
    normal_dist.append(random_value)

# Распределение результатов будет соответствовать нормальному распределению.
sns.displot(normal_dist, kind='hist')
plt.show()
```

```python
# Симуляция логнормального распределения
normal_dist = []

for _ in range(100):
    random_value = np.random.normal()
    normal_dist.append(random_value)

lognormal_dist = pd.DataFrame()
lognormal_dist['Normal'] = normal_dist
lognormal_dist['Lognormal'] = np.exp(normal_dist)
# Распределение результатов будет соответствовать логнормальному распределению.
sns.displot(lognormal_dist, kind='kde')
plt.show()
```

```python
# Симуляция распределения хи-квадрат
chi2_dist = pd.DataFrame()
df_values = [1, 3, 7, 11]

for df in df_values:
    for obs in range(1000):
        random_value = np.random.chisquare(df)
        chi2_dist.at[obs, df] = random_value

# Распределение результатов будет соответствовать распределению хи-квадрат.
sns.displot(chi2_dist, kind='kde')
plt.show()
```

```python
# Симуляция распределения Бернулли
bernoulli_dist = []

for _ in range(2):
    # Побрасываем монетку, 1 — это орел, 0 — это решка. Вероятность выпасть орлу и решке 50/50.
    trial = int(np.random.binomial(1, 0.5, 1))
    # Полученный результат добавляем в список результатов.
    bernoulli_dist.append(trial)

# Распределение результатов будет соответствовать распределению Бернулли.
sns.histplot(bernoulli_dist, bins=3)
plt.show()
```

```python
# Симуляция биномиального распределения
binomial_dist = []

for _ in range(1000):
    trials = []
    for _ in range(10):
        # Побрасываем монетку, 1 — это орел, 0 — это решка. Вероятность выпасть орлу и решке 50/50.
        trial = int(np.random.binomial(1, 0.5, 1))
        # Полученный результат добавляем в список результатов.
        trials.append(trial)
        # Считаем количество выпавших орлов в 10 бросках монетки.
        result = sum(trials)
    # Повторяем эксперимент 1000 раз. Фиксируем результат каждого эксперимента.
    binomial_dist.append(result)

# Распределение результатов будет соответствовать биномиального распределению.
sns.histplot(binomial_dist, bins=10)
plt.show()
```

```python
# Симуляция гипергеометрического распределения
hypergeometric_dist = []

for _ in range(1000):
    # Насыпаем в мешок 50 белых и 50 черных шаров.
    # Вытаскиваем 10 шаров, вытащенные шары в мешок не возвращаем.
    # Считаем количество белых шаров.
    white_balls_cnt = np.random.hypergeometric(50, 50, 10)
    # Добавляем полученный результат в список результатов. Повторяем 1000 рвз.
    hypergeometric_dist.append(white_balls_cnt)

# Распределение результатов будет соответствовать гипергеометрическому распределению.
sns.histplot(hypergeometric_dist, bins=10)
plt.show()
```

```python
# Симуляция отрицательного биномиального и геометрического распределения
neg_binomial_dist = []
# Количество успехов. Если r = 1, то мы получим геометрическое распределение.
r = 1

for _ in range(100):
    success = 0
    fail = 0
    for _ in range(10):
        # Побрасываем монетку, 1 — это орел, 0 — это решка. Вероятность выпасть орлу и решке 50/50.
        trial = int(np.random.binomial(1, 0.5, 1))
        # Если выпадает решка, то считаем попытку, как неудачную.
        if trial == 0:
            # Считаем число неуспешных попыток.
            fail += 1
        # Если выпадает орел, то считаем попытку, как удачную.
        elif trial == 1:
            # Cчитаем число успешных попыток.
            success += 1
        # Останавливаем экперимент по достижению необходимого числа успешных попыток.
        if success >= r:
            break
    # Повторяем эксперимент 1000 раз. Фиксируем количество неуспешных попыток.
    neg_binomial_dist.append(fail)

sns.histplot(neg_binomial_dist)
plt.title(f'Количество успешных попыток r={r}')
plt.show()
```

```python
# Симуляция распределения Пуассона
poisson_dist = []
# Интенсивность события, ожидаемое количество событий за фиксированное время.
lam = 3


for _ in range(100):
    # Генерируем количество событий, которые могут произойти с заданной вероятностью lam.
    event_count = np.random.poisson(lam)
    # Добавляем полученное значение в список. Повторяем 100 раз.
    poisson_dist.append(event_count)

# Строим график распределения Пуассона.
sns.displot(poisson_dist)
plt.show()
```

```python
# Симуляция экспоненциального распределения
exponential_dist = pd.DataFrame()
# Интенсивность события, ожидаемое количество событий за фиксированное время.
lam = [0.5, 1, 1.5]

for l in lam:
    for r in range(500):
        beta = 1 / l
        # Генерируем возможный временной интервал между событиями.
        period = np.random.exponential(beta)
        # Добавляем полученное значение в список. Повторяем 1000 раз
        exponential_dist.at[r, l] = period

# Строим график экспоненциального распределения.
sns.displot(exponential_dist, element='step', stat='density')
plt.show()
```

```python
# Симуляция распределения Вейбулла
weibull_dist = pd.DataFrame()
# Показатель стационарности (постоянства во времени) события.
# k=1 процесс стационарен, получаем экспоненциальное распределение.
k_values = [1, 2, 5]

for k in k_values:
    for row in range(1000):
        # Генерируем возможный временной интервал между событиями.
        period = np.random.weibull(k)
        # Добавляем полученное значение в список. Повторяем 1000 раз
        weibull_dist.at[row, k] = period

# Строим график распределения Вейбулла.
sns.kdeplot(weibull_dist, cut=-0.9)
plt.show()
```
