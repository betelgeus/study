### Шаги:
1. Сформулировать вопрос, проблему или бизнес-задачу
2. Собрать данные для анализа, обучения и тестирования
3. Подготовить и очистить данные
4. Разведывательные анализ данных, поиск закономерностей
5. Моделирование, предсказание и поиск решения проблемы
6. Визуализация, отчет и презентация шагов решения проблемы, итоговое решение
7. Поставка результата

Порядок шагов, их наличие может меняться в зависимости от задачи

### Какие задачи решаются в рамках рабочего процесса
1. Классификация. Классификация или категоризация наблюдений.
2. Корреляция.
3. Конвертация (Data Preparation).
4. Поиск и заполнение пропущенных значений
5. Коррекция. Поиск ошибок, неполных данных, выбросов и фич, которые искажают результат.
6. Генерация новых фич. Создание новых фич, которые помогают найти взаимосвязи более эффективно
7. Графики. Выбор графиков, которые лучше всего визуализируют  выводы.

### Сформулировать вопрос, проблему или бизнес-задачу

На этом этапе необходимо вместе с заказчиком сформулировать проблемы бизнеса (какую метрику хотим улучшить, какая бизнес-задача решается), которые будет решать модель. Также, требуется понять, кто участвует в проекте со стороны заказчика, кто выделяет деньги под проект, и кто принимает ключевые решения. Вдобавок необходимо узнать существуют ли готовые решения и, если да, чем они не устраивают заказчика.

Главная задача этого этапа — понять основные бизнес-переменные, которые будет прогнозировать модель. Такие переменные называются ключевыми показателями модели. После этого необходимо определить какие метрики будут использоваться, чтобы определить успешность проекта.
